# EX2: Dùng t-SNE (sklearn) giảm chiều dữ liệu MNIST về 2 chiều, so sánh chuyện giảm chiều PCA và t-SNE

```python
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mnist = fetch_openml('mnist_784', version=1, cache=True)
X = mnist.data
y = mnist.target
```

```python
X = StandardScaler().fit_transform(X)
X_ = TSNE( n_components=2, random_state=0).fit_transform(X)
```

```python
X_.shape
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x = X_[:,0], y = X_[:,1], hue = y, palette = sns.hls_palette(10), legend = 'full')
```

```python
pca = PCA(n_components=2)
pca_X = pca.fit_transform(X)
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x = pca_X[:,0], y = pca_X[:,1], hue = y, palette = sns.hls_palette(10), legend = 'full')
```
