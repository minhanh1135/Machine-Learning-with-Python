```python
dt = dt.values
```

```python
x = dt[:, 0].reshape(-1,1)
y = dt[:, 1].reshape(-1,1)
```

```python
x.shape
```

```python
plt.scatter(x, y)
```

```python
i = np.ones((x.shape[0],1))
i
```

```python
X = np.concatenate((i, x), axis = 1)
X
```

```python
X.shape
```

```python
y
```

```python
a = np.dot(X.T, X)
b = np.dot(X.T, y)
w = np.dot(np.linalg.pinv(a),b)
w
```

```python
print(X.shape, w.shape)
```

```python
plt.scatter(x, y)
plt.plot(x, np.dot(X, w))
plt.show()
```

```python
np.dot(X, w)
```

```python
def predict(x_pred, w):
  y_pred = np.dot(x_pred,w)
  return y_pred
```

```python
y_hat_50 = predict(np.array([1, 50]), w)
y_hat_50
```

```python
y_hat_100  = predict(np.array([1, 100]), w)
y_hat_100
```

```python
y_hat_150 = predict(np.array([1, 150]), w)
y_hat_150
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('/content/drive/MyDrive/ML/Machine_Learning_DSEB/Machine-Learning-with-Python/Week3-LinearRegression/housing.csv', header=None, delimiter=r"\s+", names=column_names)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
column_sels = ['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']
x_pre = data.loc[:,column_sels]
y = data[['MEDV']]
```

```python
print(x_pre.shape, y.shape)
```

```python
i = np.ones((x_pre.shape[0],1))
x_pre1 = np.concatenate((i, x_pre), axis = 1)
x_pre1
```

```python
X_norm = x_pre1.copy()
minx = np.min(x_pre1[:, 1:])
maxx = np.max(x_pre1[:, 1:])
X_norm[:, 1: ] = (x_pre1[:, 1:] - minx) / (maxx - minx)
X_norm
```

```python
y_norm = y.copy()
maxy = np.max(y)
miny= np.min(y)
y_norm = (y- miny)/ (maxy - miny)
y_norm
```

```python
a2 = np.dot(X_norm.T, X_norm)
b2 = np.dot(X_norm.T, y_norm)
w2 = np.dot(np.linalg.pinv(a2), b2)
w2
```

```python
x_draw = x_pre.values.reshape(-1, len(column_sels))
y_draw = y.values
print(x_draw.shape, y_draw.shape)
```
